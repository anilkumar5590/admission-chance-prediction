```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import data
chance = pd.read_csv('/content/Admission Chance.csv')
```

```python
# define target (y) and features (X)
y=chance['Chance of Admit ']

X=chance[['GRE Score', 'TOEFL Score', 'University Rating', ' SOP',
       'LOR ', 'CGPA', 'Research']]
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# select the linear regression model
from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
```

```python
# train or fit model
linear_regression.fit(X_train,y_train)
```

```python
linear_regression.intercept_
```

```python
linear_regression.coef_
```

```python
# predict model
y_predict = linear_regression.predict(X_test)
y_predict
```

```python
# linear regression accuracy model
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mape=mean_absolute_percentage_error(y_test,y_predict)
accuracy=100 -(mape*100)
print(f"Accuracy: {accuracy:.2f}%")
```

```python
#Linear Regression graph

plt.figure(figsize=(15,10))
plt.scatter(x=y_test, y=y_predict, color='r')
plt.plot(y_test, y_test, color='b')
plt.xlabel("Actual Value", fontsize=15)
plt.ylabel("Predicted Value", fontsize=15)
plt.title("Actual Predict vs Linear Regression Model Predict", fontsize=20)
plt.show()
```
